#### Ejercicio 1
Usando la API de Mercado Libre, obtener los ítems de una consulta de búsqueda. Pueden buscar cualquier producto de su interés.

```python
# Se importan los módulos a utilizar en los tres ejercicios

import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# URL que aparece en el navegador: 
# https://listado.mercadolibre.com.ar/apple-pencil#D[A:apple%20pencil]
    
# Al igual que en la tutorial, se busca obtener los ítems de una consulta de búsqueda, por lo que el formato a usar es el
# siguiente:
# https://api.mercadolibre.com/sites/$SITE_ID/search?q=nombre%20item

# En este caso ponemos "MLA", el ID del sitio de Argentina, en $SITE_ID. EL objeto que vamos a buscar es el Apple Pencil.

buscar = "apple pencil"
url = "https://api.mercadolibre.com/sites/{}/search?q={}".format("MLA", buscar.replace(" ", "%20"))

# Se hace el request y se guarda la response
response = requests.request("GET", url)
print(response)

# Texto
response.text
```

#### Ejercicio 2
Guarden los precios de los ítems obtenidos en un dataframe y calculen el precio promedio, el mínimo y el máximo. 

```python
# Se guardan los resultados inicialmente en un diccionario
datos_dicc = response.json()

# Se corrobora que el precio del primer resultado de la búsqueda coincide con lo que aparece en la página
print('Precio del primer resultado:', datos_dicc['results'][0]['price'])
```

```python
# Se convierten los datos en un DataFrame con el función pandas.DataFrame()
df = pd.DataFrame(datos_dicc['results'])

# Se sobreescribe el DataFrame guardando solamente las filas más relevantes
df = df[['id', 'title', 'price', 'currency_id']]

#Se chequea si algún precio no está en pesos argentinos
no_pesos = df['currency_id'] != 'ARS'
if no_pesos.any():
    print('Hay entradas cuyos precios están en otra moneda')
else:
    print('Todos los precios están en pesos argentinos')

# Se muestra el DataFrame
df
```

```python
# Se calcula el precio promedio, mínimo y máximo

precio_promedio = df['price'].mean()
precio_maximo = df['price'].max()
precio_minimo = df['price'].min()

print(f"Precio promedio: {precio_promedio}")
print(f"Precio máximo: {precio_maximo}")
print(f"Precio mínimo: {precio_minimo}")
```

#### Ejercicio 3
Armen un histograma de los precios. ¿Ven algún <i>outlier<i>? 
Nota: pueden usar la librería de Matplotlib o la de Pandas.

```python
# Se guarda la columna de precios en una serie
precios = df['price']

# Se configura el tamaño del gráfico
plt.figure(figsize=(13, 9))

# Se crea el histograma
plt.hist(df['price'], bins=20, color='lightblue', edgecolor='black')

# Se agregan etiquetas y título
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Histograma de Precios')

# Se muestra el histograma
plt.show()
```
